 # Pengaruh Tipe Kamar Terhadap Harga


```python
# Menghitung harga rata-rata berdasarkan tipe ruangan
room_type_mean_prices = df.groupby('room_type')['price'].mean().reset_index()

# Mengurutkan berdasarkan harga
sorted_room_type_mean_prices = room_type_mean_prices.sort_values(by='price', ascending=False)
sorted_room_type_mean_prices.reset_index(drop=True, inplace=True)

# Menampilkan hasilnya
sorted_room_type_mean_prices
```

```python
# Fungsi untuk membungkus teks menjadi beberapa baris
def wrap_text(text):
    return '\n'.join(wrap(text, 12))

# Plot the bar chart of mean prices by room type
plt.figure(figsize=(6, 4))
sns.barplot(data=sorted_room_type_mean_prices, x='room_type', y='price', palette='Blues_r')
plt.xlabel('Room Type')
plt.ylabel('Average Price ($)')
plt.title('The Effect of Room Type on Average Price in Singapore')

# Ganti tick label sumbu x dengan versi split baris
plt.xticks(range(len(sorted_room_type_mean_prices['room_type'])), [wrap_text(tick) for tick in sorted_room_type_mean_prices['room_type']], rotation=0)

# Tunjukkan plot
plt.show()
```

Interpretasi Plot berdasarkan barchar berikut ialah:

1. **Private Room**: Dengan harga rata-rata \$133.50, tipe ruangan ini adalah yang paling mahal di antara semua tipe yang tersedia. Hal ini mungkin mencerminkan tingkat privasi dan fasilitas tambahan yang ditawarkan oleh tipe ruangan ini.

2. **Hotel Room**: Tipe ruangan ini memiliki harga rata-rata \$116.32, menempatkannya di urutan kedua dalam hal harga. Ini mungkin mencakup kamar-kamar hotel yang menawarkan layanan dan fasilitas yang lebih standar dibandingkan dengan kamar pribadi.

3. **Entire Home/Apartment**: Meskipun menawarkan seluruh rumah atau apartemen, tipe ruangan ini memiliki harga rata-rata \$111.08, yang lebih rendah daripada kamar pribadi dan hotel. Ini mungkin menunjukkan bahwa tipe ruangan ini lebih fokus pada penyewaan jangka panjang atau mungkin lokasi yang kurang strategis.

4. **Shared Room**: Sebagai tipe ruangan dengan harga terendah, kamar bersama memiliki harga rata-rata \$50.91. Tipe ruangan ini mungkin menarik bagi pelancong dengan anggaran terbatas yang tidak memerlukan banyak privasi atau fasilitas tambahan.

Secara keseluruhan, plot dan tabel ini menunjukkan variasi harga berdasarkan tipe ruangan, mencerminkan perbedaan dalam tingkat privasi, fasilitas, dan mungkin lokasi dan tujuan penyewaan. Interpretasi ini dapat diperdalam dengan mempertimbangkan faktor-faktor lain seperti lokasi, ulasan, atau ketersediaan.

# Trend Penyewaan dari 01-01-2018 hingga 22-09-2022

```python
# Replace the DataFrame if you have a specific one for Singapore
singapore_df = df

# Convert the date column to datetime
singapore_df['date'] = pd.to_datetime(singapore_df['date'])

# Filter the data for the specified date range
start_date = '2018-01-01'
end_date = '2022-09-22'
mask = (singapore_df['date'] >= start_date) & (singapore_df['date'] <= end_date)
filtered_df = singapore_df[mask]

# Get the number of rentals per month
# Replace 'listing_id' with the relevant column in your dataset
rentals_per_month = filtered_df.resample('M', on='date')['listing_id'].count().reset_index()

# Plot the number of rentals per month
plt.figure(figsize=(15, 6))
sns.lineplot(data=rentals_per_month, x='date', y='listing_id', marker='o', markersize=5)
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penyewaan')
plt.title('Rental Trends from January 1, 2018 to September 22, 2022')
plt.show()
```

Interpretasi berdasarkan trend waktu dari 1 Januari 2018 hingga 22 September 2022 ialah pada tahun 2018 awal hingga 2020 awal terjadi kenaikan jumlah penyewaan yang cukup stabil dimana pada waktu-waktu tersebut cukup banyak tamu yang memesan tipe-tipe kamar di AirBnb Singapore.
```python
# Menghitung rata-rata jumlah ulasan per neighbourhood_group
rata_rata_ulasan_per_grup = df.groupby('neighbourhood_group')['number_of_reviews'].mean()

# Fungsi yang menandai listing dengan jumlah ulasan di atas rata-rata grupnya
def di_atas_rata_rata(row, rata_rata_ulasan):
    return row['number_of_reviews'] > rata_rata_ulasan[row['neighbourhood_group']]

# Menandai listing dengan jumlah ulasan di atas rata-rata grupnya
df['di_atas_rata_rata'] = df.apply(lambda row: di_atas_rata_rata(row, rata_rata_ulasan_per_grup), axis=1)

# Mengambil listing dengan jumlah ulasan di atas rata-rata untuk setiap neighbourhood_group
listing_di_atas_rata_rata = df[df['di_atas_rata_rata']]

# Mengelompokkan listing berdasarkan neighbourhood_group dan menyimpan daftar nama yang dihitung
listing_per_grup = listing_di_atas_rata_rata.groupby('neighbourhood_group')['name'].apply(list)

# Menggabungkan DataFrame dan menampilkan hasil dalam format tabel
hasil = listing_di_atas_rata_rata[['neighbourhood_group', 'name', 'number_of_reviews']].copy()
hasil.reset_index(drop=True, inplace=True)
hasil
```

```python
from geopy.distance import distance

# Set the coordinates of the coastal point
coastal_point = (1.3122, 103.8764)

# Calculate the distance between each listing and the coastal point
listings_in_neighbourhood['distance_to_coast'] = listings_in_neighbourhood.apply(
    lambda row: distance((row.latitude, row.longitude), coastal_point).km, axis=1)

# Find the closest listing to the coastal point
closest_listing = listings_in_neighbourhood.loc[listings_in_neighbourhood['distance_to_coast'].idxmin()]

# Print the distance and the ID of the closest listing
print(f"The closest distance from the most active listings in {most_active_neighbourhood} to the coast is "
      f"{closest_listing['distance_to_coast']:.2f} km, dengan ID listing {closest_listing['id']}.")
```

```python
# Ekstrak tahun dari kolom 'date' dan buat kolom baru 'year'
df["year"] = df["date"].dt.year

# Filter data dari tahun 2018 hingga 2022
df = df[(df["year"] >= 2018) & (df["year"] <= 2022)]

# Kelompokkan data berdasarkan tahun dan hitung jumlah review
review_per_tahun = df.groupby("year")["number_of_reviews"].sum()

# Menghitung persentase akumulasi jumlah reviews
review_cumsum = review_per_tahun.cumsum()
total_reviews = review_per_tahun.sum()
percentage_accumulation = (review_cumsum / total_reviews) * 100

# Menghitung kenaikan dan penurunan paling tajam
yearly_changes = review_per_tahun.pct_change()
sharpest_increase = yearly_changes.max()
sharpest_decrease = yearly_changes.min()

print(f"The sharpest increase: {sharpest_increase * 100}%")
print(f"The sharpest decrease: {sharpest_decrease * 100}%")

# Membuat grafik jumlah review per tahun
fig, ax = plt.subplots()
review_per_tahun.plot(kind="bar", ax=ax)
ax.set_title("Number of Reviews per Year")
ax.set_ylabel("Reviews")
ax.set_xticklabels(review_per_tahun.index, rotation=0)
plt.show()
```

```python
# Membuat grafik persentase akumulasi jumlah reviews
fig, ax = plt.subplots()
percentage_accumulation.plot(kind="bar", ax=ax)
ax.set_title("Percentage of Accumulated Number of Reviews")
ax.set_ylabel("Percentage %")
ax.set_ylim([0, 100])
ax.set_xticklabels(percentage_accumulation.index, rotation=0)
plt.show()
```
